```python
#Removing the df2 header as from above the header row represent employee data

df2 = pd.read_csv(r"C:\Users\USER\Desktop\Python\Python Exercises\2024\Data\DM_2.csv", header=None)
```

```python
#Updating the column names

df2.columns = ['First name', 'Last name', 'Days active']

df2
```

```python
#Editing the last name column because the names are in all caps 

list2 = []
for name in df2['Last name']:
    list2.append(name.title())
```

```python
#Update the data frame

df2['Last name'] = list2
```

```python
df2
```

```python
#We shalll remove the repetead entries and only remain with the entry with the highest days active
#Repeated entries - Zane and Elvis

df2.drop(index=[5,7,8], inplace = True)

df2
```

```python
#The final df2 will be the first and last name combined.

df2['Name'] = df2['First name'] + " " + df2['Last name']

df2.drop(columns=['First name','Last name'], inplace=True)

df2
```

```python
#combining the two data frames.

df = pd.merge(df1,df2,how='left',on='Name')

df
```

```python
#Employee name with the highest salary

df[df['Salary'] == df['Salary'].max()][['Name','Salary']]
```

```python
#Details of the employee with the lowest salary

df[df['Salary'] == df['Salary'].min()]
```
